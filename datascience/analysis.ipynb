```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

part 1 : Loading and processing the data

```python
pd.set_option('display.max_columns',None)
```

```python
canada = pd.read_excel('Canada.xlsx',
               sheet_name='Canada by Citizenship',
               skiprows=20,
               skipfooter=2)
```

```python
canada.head()
```

```python
canada.info()
```

```python
canada.describe()
```

```python
canada.describe(include='object')
```

```python
canada.columns = canada.columns.astype(str)  # convert column names to strings
```

```python
print(canada.columns.tolist())
```

```python
cols_to_drops = ['Type','Coverage','AREA','REG','DEV']
canada.drop(columns=cols_to_drops, inplace=True)
```

```python
canada.head(2)
```

Add columns

```python
canada.shape
```

```python
years = [ '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']
```

```python
canada['total']=canada.loc[:,years].sum(axis=1)
canada.head(3)
```

```python
canada['mean']=canada.loc[:,years].sum(axis=1)
canada.head(3)
```

```python
canada['faltu'] = 0
canada.head(3)
```

```python
canada.rename(
    columns = {
        'OdName': 'Country',
        'AreaName': 'Continent',
        'RegName': 'Region',
        'DevName': 'Status',
    },
    inplace=True 
)
```

```python
canada.head()
```

```python
canada.set_index('Country', inplace = True)
```

```python
canada.loc[['Japan','France','India','China','Philippines'],years].T.plot(kind='line',figsize=(15,6),style = 'x--')
```

```python
canada.loc[['Japan','France','India',],years].plot(kind='bar',figsize=(15,6),style = 'x--')
```

```python
cols_to_drops = ['1980-1990']
canada.drop(columns=cols_to_drops, inplace=True)
```

```python
canada.head()
```

```python
cols_to_drops = ['faltu']
canada.drop(columns=cols_to_drops, inplace=True)
```

```python
canada.head()
```

```python
canada['1980-1990']=canada.loc[:,map(str,range(1980,1990))].sum(axis=1)
canada['1990-1999']=canada.loc[:,map(str,range(1990,1999))].sum(axis=1)
canada['2000-2009']=canada.loc[:,map(str,range(2000,2009))].sum(axis=1)
canada.head()
```

```python
# fix the country name 
print(canada.index.tolist())
```

```python
decade_cols = ['1980-1990','1990-1999','2000-2009']
plt.figure(facecolor='white')
canada.loc['Japan',decade_cols].plot(kind='pie',figsize=(10,6), autopct='%1.1f%%')
```

```python
country_list= canada.index.tolist()
```

```python
canada.loc[country_list,decade_cols].T
```

```python
canada.loc[country_list,decade_cols].T.plot(
    kind='pie',
    figsize=(10,6), 
    autopct='%1.1f%%',
    subplots=True)
```

```python
canada.loc[country_list[:10],decade_cols].T.plot(
    kind='pie',
    figsize=(50,10), 
    autopct='%1.1f%%',
    subplots=True,
    layout=(2,5),
    )
plt.show()
plt.savefig('first five countries.png',bbox_inches='tight')
```

```python
canada.loc[country_list[:10],decade_cols].T.plot(
    kind='bar',
    figsize=(50,10), 
    subplots=True,
    layout=(2,5),
    )
plt.show()
```
