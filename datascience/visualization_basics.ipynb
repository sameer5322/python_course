```python
x = np.arange(-10, 10, 0.1)
print(x.shape)
```

```python
x2 = x ** 2 # square of x
```

```python
_ = plt.plot(x,x2)
```

```python
x3 = x ** 3
```

```python
plt.figure(figsize = (10,10)) # create a figure with a size of 10X10
plt.plot(x, label = 'original')
plt.plot(x2, label = 'squared')
plt.plot(x3, label = 'cubed')
plt.legend() # activate the legends  
plt.xlabel("Data Points") # add a label to x-axis
plt.ylabel("Values of X") # add a label to y-axis
plt.title('The visualization of X and its square and cube') # add title to the figure
plt.show() # show the figure
```

```python
cities = ['lucknow','delhi','mumbai','pune','banglore']
rainfall = [39,20,62,45,78]
```

```python
plt.bar(cities ,rainfall)
```

```python
plt.barh(cities ,rainfall)
```

```python
plt.figure(figsize=(10,7))
plt.bar(cities ,rainfall , color='red', edgecolor='black', hatch='//', width=[.1,.2,.3,.5,.2], align='edge')
plt.xlabel("Cities") # add a label to x-axis
plt.ylabel("Rainfall") # add a label to y-axis
plt.title('rainfall chart') # add title to the figure
```

```python
plt.figure(facecolor='white')
plt.pie(rainfall, labels=cities, autopct='%1.1f%%', radius=2)
plt.show()
```

```python
plt.figure(facecolor='white')
wedges = [0,.3,0,0,0]
plt.pie(rainfall, labels=cities, autopct='%1.1f%%', radius=2, explode=wedges)
plt.show()
```
